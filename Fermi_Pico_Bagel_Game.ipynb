```python
Rules='''
   Rules:
         Fermi - A Correct Digit Placed Correctly
         Pico - A Correct Digit Placed Incorrectly
         Bagels - No Digits Are Correct'''
while True:
  print('''\n*******Welcome to The Fermi Pico Bagel Game*******
              
        -------Number Guessing Game-------
                    1.Play Game
                    2.Rules
                    3.Quit
                      ''')
  original_number_list=[]
  import random
  choice=input('Enter Your Choice : ')
  if choice == '1':

    try:
      no=int(input('How many Digit You Want in number : '))
    except ValueError as e:
      print('Please Enter a Integers Only')
      continue
    random_number=random.randint(0,9)
    for i in range(no):
      while random_number in original_number_list:
        random_number=random.randint(0,9)
      original_number_list.append(random_number)
    
      

    if original_number_list[0] == 0:
      original_number_list[0],original_number_list[1]=original_number_list[1],original_number_list[0]


    while True:
      try:
        guess_number=int(input("Enter a Number or If You Want to Quit just press 'q' : "))
      except ValueError as e:
        print('Please Enter a Integers Only')
        continue
      if guess_number == 'q':
        break
      else:
        gn_list=[]
        while guess_number > 0:
          gn_list.append(guess_number % 10)
          guess_number = guess_number // 10
        guess_number_list=gn_list[::-1]
        if len(guess_number_list) != len(original_number_list):
          guess_number
          continue
        elif len(set(guess_number_list)) != len(guess_number_list):
          print('duplicate number')
          continue
        elif guess_number_list == original_number_list:
          print('Fermi'*len(original_number_list))
          print('You Win !!')
          break
        output=[]
        for i in range(len(original_number_list)):
          if original_number_list[i] == guess_number_list[i]:
            output.append('Fermi')
          elif original_number_list[i] in guess_number_list:
            output.append('Pico')
        output_string=''
        for i in output:
          output_string+=i + ' '
        if len(output) == 0:
          print('Bagel')
        else:
          print(output_string)
  elif choice == '2':
    print(Rules)
  elif choice == '3':
    break
```
